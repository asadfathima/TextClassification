```python
# generate synthetic data
import json
import pandas as pd
import time
import openai
import openai

openai.api_key = ''

# Load your training data
new_train_df = pd.read_csv('/Users/asadfathima/Desktop/Assignment-AI/text-classification-with-llms/train.csv')

# Calculate label counts
label_counts = new_train_df['label_text'].value_counts().reset_index()
label_counts.columns = ['label_text', 'count']

needed_rows = 50

# Function to make an API call with retries
def make_api_call_with_retries(query, max_retries=10, sleep_time=2):
    retries = 0
    while retries < max_retries:
        try:
            # Replace this with the actual API call code
            response = openai.ChatCompletion.create(
                model="gpt-3.5-turbo-0125",
                messages=[
                    {"role": "system", "content": "You are a helpful assistant that generates synthetic training data."},
                    {"role": "user", "content": query},
                ],
                temperature=0,
                response_format={"type": "json_object"},
            )
            print(response.choices[0].message.content)
            resp = json.loads(response.choices[0].message.content)
            resp = list(resp.values())[0]
            return resp
        except Exception as e:
            print(f"API call failed, retrying... ({e})")
            time.sleep(sleep_time)
            retries += 1

for index, row in label_counts.iterrows():
    if row['count'] < needed_rows:
        n_rows_to_generate = needed_rows - row['count']
        df_rows_with_label = new_train_df.loc[new_train_df['label_text'] == row['label_text']]
        training_rows = [row_label['text'] for _, row_label in df_rows_with_label.iterrows()]

        query = f"""This is an array of some texts with label "{row['label_text']}".
        {json.dumps(training_rows)}
        Generate {n_rows_to_generate} more similar texts for my training data, return a json array of strings and nothing else.
        """

        # Using the function with retries to make the API call
        response = make_api_call_with_retries(query)
        print(response)

        for new_row_to_add in response:
            max_id = new_train_df['id'].max() if not new_train_df.empty else 0
            new_id = int(max_id) + 1
            new_row = {'id': new_id, 'text': new_row_to_add, 'label_text': row['label_text']}
            new_train_df.loc[len(new_train_df)] = new_row
```

```python
import time
import random

MODEL = fine_tuned_model


def ask_with_retry(query: str, model: str = MODEL, max_attempts: int = 3) -> str:
    base_delay = 1  # base delay in seconds
    for attempt in range(max_attempts):
        try:
            response = client.chat.completions.create(
                model=model,
                messages=[
                    {"role": "system", "content": "You are ClassiFine"},
                    {"role": "user", "content": query},
                ],
                temperature=0,
            )
            return response.choices[0].message.content.strip()
        except Exception as e:
            if attempt < max_attempts - 1:
                # Calculate sleep time using exponential backoff with full jitter
                sleep_time = base_delay * 2 ** attempt
                sleep_time = sleep_time * random.uniform(0.5, 1.5)
                print(f"Attempt {attempt + 1} failed, retrying in {sleep_time:.2f} seconds...")
                time.sleep(sleep_time)
            else:
                print(f"All attempts failed. Last error message: {e}")
                return "not toxic"
```

```python
import pandas as pd
test_file_path = '/Users/asadfathima/Desktop/Assignment-AI/text-classification-with-llms/test.csv'
test_df = pd.read_csv(test_file_path)

output_data = []

for row in test_df.iterrows():
    predicted_label = ask_with_retry(row[1].text)
    id = row[1].id
    print(f"{id} predicted_label: {predicted_label}")
    output_data.append({
        "id": id,
        "label_text": predicted_label
    })


output_df = pd.DataFrame(output_data)
```
